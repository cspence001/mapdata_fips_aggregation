<b>State Aggregation

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
va_hpgf_state = va_hpgf.groupby(['FIPSstate'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum','ForgivenessAmount':'sum'})
va_hpgf_state[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']] = va_hpgf_state[['CurrentApprovalAmount', 'Estimated_LenderProfit', 'ForgivenessAmount']].apply(lambda x: round(x, 2)) #format

va_hpgf_state
```

<h4>County Aggregation 

```python
#cntyloc = va_hpgf.loc[va_hpgf["FIPScounty"] == ""
```

```python
va_hpgf_cnty = va_hpgf.groupby(['FIPScounty'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum', 'ForgivenessAmount':'sum'})
va_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']] = va_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']].astype(float) # object > float
va_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']] = va_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']].apply(lambda x: round(x, 2)) #format
va_hpgf_cnty.head()#132
```

```python
va_hpgf_cnty["CurrentApprovalAmount"].max()
```

```python
#NonProfit Reporting
pd.set_option('display.max_columns', None)
nonprofit = va_hpgf.loc[va_hpgf["NonProfit"] == "Y"]
nonprofit #1536

va_hpgf_cnty_nonprofit = nonprofit.groupby(['FIPScounty'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum', 'ForgivenessAmount':'sum'})
va_hpgf_cnty_nonprofit["NonProfitLoans"] = va_hpgf_cnty_nonprofit.LoanNumber
va_hpgf_cnty_nonprofit["NonProfitApprovalAmount"] = va_hpgf_cnty_nonprofit.CurrentApprovalAmount
va_hpgf_cnty_nonprofit["NonProfitJobsReported"] = va_hpgf_cnty_nonprofit.JobsReported
va_hpgf_cnty_nonprofit["NonProfitLenderProfit"] = va_hpgf_cnty_nonprofit.Estimated_LenderProfit
cols = ['FIPScounty', 'NonProfitLoans', 'NonProfitApprovalAmount', 'NonProfitJobsReported', 'NonProfitLenderProfit']
va_hpgf_cnty_nonprofit = va_hpgf_cnty_nonprofit[cols]
va_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']] = va_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']].apply(lambda x: round(x, 2)) #format
#va_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']] = va_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']].astype(float) # object > float
va_hpgf_cnty_nonprofit
```
