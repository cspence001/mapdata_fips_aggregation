<b>State Aggregation

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
tx_hpgf_state = tx_hpgf.groupby(['FIPSstate'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum','ForgivenessAmount':'sum'})
tx_hpgf_state[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']] = tx_hpgf_state[['CurrentApprovalAmount', 'Estimated_LenderProfit', 'ForgivenessAmount']].apply(lambda x: round(x, 2)) #format

tx_hpgf_state
```

<h4>County Aggregation 

```python
#cntyloc = tx_hpgf.loc[tx_hpgf["FIPScounty"] == ""
```

```python
tx_hpgf_cnty = tx_hpgf.groupby(['FIPScounty'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum', 'ForgivenessAmount':'sum'})
tx_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']] = tx_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']].astype(float) # object > float
tx_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']] = tx_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']].apply(lambda x: round(x, 2)) #format
tx_hpgf_cnty.head()#132
```

```python
tx_hpgf_cnty["CurrentApprovalAmount"].max()
```

```python
#NonProfit Reporting
pd.set_option('display.max_columns', None)
nonprofit = tx_hpgf.loc[tx_hpgf["NonProfit"] == "Y"]
nonprofit #1536

tx_hpgf_cnty_nonprofit = nonprofit.groupby(['FIPScounty'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum', 'ForgivenessAmount':'sum'})
tx_hpgf_cnty_nonprofit["NonProfitLoans"] = tx_hpgf_cnty_nonprofit.LoanNumber
tx_hpgf_cnty_nonprofit["NonProfitApprovalAmount"] = tx_hpgf_cnty_nonprofit.CurrentApprovalAmount
tx_hpgf_cnty_nonprofit["NonProfitJobsReported"] = tx_hpgf_cnty_nonprofit.JobsReported
tx_hpgf_cnty_nonprofit["NonProfitLenderProfit"] = tx_hpgf_cnty_nonprofit.Estimated_LenderProfit
cols = ['FIPScounty', 'NonProfitLoans', 'NonProfitApprovalAmount', 'NonProfitJobsReported', 'NonProfitLenderProfit']
tx_hpgf_cnty_nonprofit = tx_hpgf_cnty_nonprofit[cols]
tx_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']] = tx_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']].apply(lambda x: round(x, 2)) #format
#tx_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']] = tx_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']].astype(float) # object > float
tx_hpgf_cnty_nonprofit
```
