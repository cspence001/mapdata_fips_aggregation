<b>State Aggregation

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
il_hpgf_state = il_hpgf.groupby(['FIPSstate'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum','ForgivenessAmount':'sum'})
il_hpgf_state[['CurrentApprovalAmount','Estimated_LenderProfit']] = il_hpgf_state[['CurrentApprovalAmount','Estimated_LenderProfit']].astype(float) # object > float
il_hpgf_state[['CurrentApprovalAmount','Estimated_LenderProfit']] = il_hpgf_state[['CurrentApprovalAmount','Estimated_LenderProfit']].apply(lambda x: round(x, 2)) #format
il_hpgf_state
```

<h4>County Aggregation 

```python
il_hpgf_cnty = il_hpgf.groupby(['FIPScounty'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum', 'ForgivenessAmount':'sum'})
il_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']] = il_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']].astype(float)#float
il_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']] = il_hpgf_cnty[['CurrentApprovalAmount','Estimated_LenderProfit', 'ForgivenessAmount']].apply(lambda x: round(x, 2)) # format
il_hpgf_cnty.head()#132
```

```python
#NonProfit Reporting
pd.set_option('display.max_columns', None)
nonprofit = il_hpgf.loc[il_hpgf["NonProfit"] == "Y"]
nonprofit #1536

il_hpgf_cnty_nonprofit = nonprofit.groupby(['FIPScounty'],as_index=False).agg({'LoanNumber': 'count', 'CurrentApprovalAmount':'sum', 'JobsReported':'sum', 'ServicingLenderLocationID': 'nunique', 'Estimated_LenderProfit':'sum', 'ForgivenessAmount':'sum'})
il_hpgf_cnty_nonprofit["NonProfitLoans"] = il_hpgf_cnty_nonprofit.LoanNumber
il_hpgf_cnty_nonprofit["NonProfitApprovalAmount"] = il_hpgf_cnty_nonprofit.CurrentApprovalAmount
il_hpgf_cnty_nonprofit["NonProfitJobsReported"] = il_hpgf_cnty_nonprofit.JobsReported
il_hpgf_cnty_nonprofit["NonProfitLenderProfit"] = il_hpgf_cnty_nonprofit.Estimated_LenderProfit
cols = ['FIPScounty', 'NonProfitLoans', 'NonProfitApprovalAmount', 'NonProfitJobsReported', 'NonProfitLenderProfit']
il_hpgf_cnty_nonprofit = il_hpgf_cnty_nonprofit[cols]
il_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']] = il_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']].apply(lambda x: round(x, 2)) #format
#il_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']] = il_hpgf_cnty_nonprofit[['NonProfitApprovalAmount','NonProfitLenderProfit']].astype(float) # object > float
il_hpgf_cnty_nonprofit
```
